```python
#Create a Dataframe having year wise details of average humidity,temp,wind chill, wind mph
df_with_YearWise_AvgData = pd.read_sql_query("SELECT season as Year,avg(humidity) as avg_Humidity,avg(temp) as avg_Temp,avg(wind_chill) as avg_Wind_Chill,avg(wind_mph) as avg_Wind_Mph FROM 'game_info' group by season", con)
df_with_YearWise_AvgData
```

```python
#Create a dataframe haviing year wise detail of no. of time home team has won when various criterias are below the average humidity,temp,wind chill and wind mph of that year
df_homeTeam_winsWith_belowAvg_Humidity = pd.read_sql_query("SELECT season as Year,count(home) as homeTeam_winsWith_belowAvg_Humidity FROM 'game_info' where humidity < (select avg(humidity) from 'game_info') and home = winner group by season", con)
df_homeTeam_winsWith_belowAvg_Temp = pd.read_sql_query("SELECT season as Year,count(home) as homeTeam_winsWith_belowAvg_Temp FROM 'game_info' where temp < (select avg(temp) from 'game_info') and home = winner group by season", con)
df_homeTeam_winsWith_belowAvg_WindChill = pd.read_sql_query("SELECT season as Year,count(home) as homeTeam_winsWith_belowAvg_WindChill FROM 'game_info' where wind_chill < (select avg(wind_chill) from 'game_info') and home = winner group by season", con)
df_homeTeam_winsWith_belowAvg_WindMph = pd.read_sql_query("SELECT season as Year,count(home) as homeTeam_winsWith_belowAvg_WindMph FROM 'game_info' where wind_mph < (select avg(wind_mph) from 'game_info') and home = winner group by season", con)
```

```python
#Create a dataframe haviing year wise detail of no. of time home team has won when various criterias are above the average humidity,temp,wind chill and wind mph of that year
df_homeTeam_winsWith_aboveAvg_Humidity = pd.read_sql_query("SELECT season as Year,count(home) as homeTeam_winsWith_aboveAvg_Humidity FROM 'game_info' where humidity > (select avg(humidity) from 'game_info') and home = winner group by season", con)
df_homeTeam_winsWith_aboveAvg_Temp = pd.read_sql_query("SELECT season as Year,count(home) as homeTeam_winsWith_aboveAvg_Temp FROM 'game_info' where temp > (select avg(temp) from 'game_info') and home = winner group by season", con)
df_homeTeam_winsWith_aboveAvg_WindChill = pd.read_sql_query("SELECT season as Year,count(home) as homeTeam_winsWith_aboveAvg_WindChill FROM 'game_info' where wind_chill > (select avg(wind_chill) from 'game_info') and home = winner group by season", con)
df_homeTeam_winsWith_aboveAvg_WindMph = pd.read_sql_query("SELECT season as Year,count(home) as homeTeam_winsWith_aboveAvg_WindMph FROM 'game_info' where wind_mph > (select avg(wind_mph) from 'game_info') and home = winner group by season", con)
```

```python
#Create a dataframe haviing year wise detail of no. of time away team has won when various criterias are below the average humidity,temp,wind chill and wind mph of that year
df_awayTeam_winsWith_belowAvg_Humidity = pd.read_sql_query("SELECT season as Year,count(away) as awayTeam_winsWith_belowAvg_Humidity FROM 'game_info' where humidity < (select avg(humidity) from 'game_info') and away = winner group by season", con)
df_awayTeam_winsWith_belowAvg_Temp = pd.read_sql_query("SELECT season as Year,count(away) as awayTeam_winsWith_belowAvg_Temp FROM 'game_info' where temp < (select avg(temp) from 'game_info') and away = winner group by season", con)
df_awayTeam_winsWith_belowAvg_WindChill = pd.read_sql_query("SELECT season as Year,count(away) as awayTeam_winsWith_belowAvg_WindChill FROM 'game_info' where wind_chill < (select avg(wind_chill) from 'game_info') and away = winner group by season", con)
df_awayTeam_winsWith_belowAvg_WindMph = pd.read_sql_query("SELECT season as Year,count(away) as awayTeam_winsWith_belowAvg_WindMph FROM 'game_info' where wind_mph < (select avg(wind_mph) from 'game_info') and away = winner group by season", con)
```

```python
#Create a dataframe haviing year wise detail of no. of time away team has won when various criterias are above the average humidity,temp,wind chill and wind mph of that year
df_awayTeam_winsWith_aboveAvg_Humidity = pd.read_sql_query("SELECT season as Year,count(away) as awayTeam_winsWith_aboveAvg_Humidity FROM 'game_info' where humidity > (select avg(humidity) from 'game_info') and away = winner group by season", con)
df_awayTeam_winsWith_aboveAvg_Temp = pd.read_sql_query("SELECT season as Year,count(away) as awayTeam_winsWith_aboveAvg_Temp FROM 'game_info' where temp > (select avg(temp) from 'game_info') and away = winner group by season", con)
df_awayTeam_winsWith_aboveAvg_WindChill = pd.read_sql_query("SELECT season as Year,count(away) as awayTeam_winsWith_aboveAvg_WindChill FROM 'game_info' where wind_chill > (select avg(wind_chill) from 'game_info') and away = winner group by season", con)
df_awayTeam_winsWith_aboveAvg_WindMph = pd.read_sql_query("SELECT season as Year,count(away) as awayTeam_winsWith_aboveAvg_WindMph FROM 'game_info' where wind_mph > (select avg(wind_mph) from 'game_info') and away = winner group by season", con)
```

```python
#Merging the average value dataframe with the count of home and away winners wih respect to below or above average humidity,temp etc.
df_forBar_Chart = pd.concat([df_with_YearWise_AvgData, df_homeTeam_winsWith_belowAvg_Humidity["homeTeam_winsWith_belowAvg_Humidity"], df_homeTeam_winsWith_belowAvg_Temp["homeTeam_winsWith_belowAvg_Temp"], df_homeTeam_winsWith_belowAvg_WindChill["homeTeam_winsWith_belowAvg_WindChill"], df_homeTeam_winsWith_belowAvg_WindMph["homeTeam_winsWith_belowAvg_WindMph"], df_homeTeam_winsWith_aboveAvg_Humidity["homeTeam_winsWith_aboveAvg_Humidity"], df_homeTeam_winsWith_aboveAvg_Temp["homeTeam_winsWith_aboveAvg_Temp"], df_homeTeam_winsWith_aboveAvg_WindChill["homeTeam_winsWith_aboveAvg_WindChill"], df_homeTeam_winsWith_aboveAvg_WindMph["homeTeam_winsWith_aboveAvg_WindMph"], df_awayTeam_winsWith_belowAvg_Humidity["awayTeam_winsWith_belowAvg_Humidity"], df_awayTeam_winsWith_belowAvg_Temp["awayTeam_winsWith_belowAvg_Temp"], df_awayTeam_winsWith_belowAvg_WindChill["awayTeam_winsWith_belowAvg_WindChill"], df_awayTeam_winsWith_belowAvg_WindMph["awayTeam_winsWith_belowAvg_WindMph"], df_awayTeam_winsWith_aboveAvg_Humidity["awayTeam_winsWith_aboveAvg_Humidity"], df_awayTeam_winsWith_aboveAvg_Temp["awayTeam_winsWith_aboveAvg_Temp"], df_awayTeam_winsWith_aboveAvg_WindChill["awayTeam_winsWith_aboveAvg_WindChill"], df_awayTeam_winsWith_aboveAvg_WindMph["awayTeam_winsWith_aboveAvg_WindMph"]], axis=1)
df_forBar_Chart.fillna(0, inplace=True)
pd.options.display.float_format = '{:,.0f}'.format
df_forBar_Chart.set_index('Year')
df_forBar_Chart[['Year','avg_Humidity','avg_Temp','avg_Wind_Chill','avg_Wind_Mph','homeTeam_winsWith_belowAvg_Humidity','homeTeam_winsWith_belowAvg_Temp','homeTeam_winsWith_belowAvg_WindChill','homeTeam_winsWith_belowAvg_WindMph','homeTeam_winsWith_aboveAvg_Humidity','homeTeam_winsWith_aboveAvg_Temp','homeTeam_winsWith_aboveAvg_WindChill','homeTeam_winsWith_aboveAvg_WindMph']]
df_forBar_Chart 
```

```python
# Adding lart row in the dataframe with aggregate values an saving the dataframe to sqlite db
import numpy as np
total = df_forBar_Chart.apply(np.sum)
total['Year'] = 'Aggregate'
total['avg_Humidity'] = total['avg_Humidity'] / 17
total['avg_Temp'] = total['avg_Temp'] / 17
total['avg_Wind_Chill'] = total['avg_Wind_Chill'] / 17
total['avg_Wind_Mph'] = total['avg_Wind_Mph'] / 17
total = pd.DataFrame(total.values, index=total.keys()).T
df_forBar_Chart = total.append(df_forBar_Chart, ignore_index=True)
df_forBar_Chart
df_forBar_Chart.to_sql("dataForBarGraph", con, if_exists="replace")
```
